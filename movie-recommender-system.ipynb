## POPULARITY BASED RECOMD SYS

```python
#merging Two datasets on common coloumn movieID
rating_with_movie = rating.merge(movie,on='movieId').merge(links,on='movieId')
```

```python
rating_with_movie.drop(['index','timestamp'],axis=1,inplace=True)
```

```python
rating_with_movie.shape
```

```python
rating_with_movie = rating_with_movie.merge(poster,on='imdbId')
rating_with_movie.drop(['Movie-Name','Movie-Genres','tmdbId'], axis=1)
```

```python
num_rating_df = rating_with_movie.groupby(['Movie-Name','imdbId','tmdbId']).count()[["rating"]].reset_index()
num_rating_df.rename(columns={'rating':'Movie-Rating'},inplace = True) 
```

```python
num_rating_df
```

```python
avg_rating_df = rating_with_movie.groupby(['Movie-Name']).mean()["rating"].reset_index()
avg_rating_df.rename(columns={'rating':'AVG-Rating'},inplace = True) 
avg_rating_df.head()
```

```python
popular_df = num_rating_df.merge(avg_rating_df, on='Movie-Name')
popular_df
```

```python
popular_df.describe()
```

```python
#need the data with movie rating more than 100
#filtering the data
popular_df_rated = popular_df[popular_df['Movie-Rating']>=80].sort_values('AVG-Rating' , ascending = False).head(30)
popular_df_rated.merge(movie,on=['Movie-Name'])
```

```python
popular_df_rated
```

## Collabrative Filtering

```python
#filtering out the users who rated around 100 movies i,e its and experienced user
x = rating_with_movie.groupby('userId').count()['rating']>90
bade_log= x[x].index #taking out the index and filterinf out false values
```

```python
#creating experienced user dataset which rated more than the set criteria that is 100 
filtered_rating= rating_with_movie[rating_with_movie['userId'].isin(bade_log)] #filtering the rating_with_movie with rmoving user expect in bade_log
```

```python
filtered_rating
```

```python
y = filtered_rating.groupby('Movie-Name').count()['rating']>=38
famous_movies = y[y].index
famous_movies
```

```python
final_rating = filtered_rating[filtered_rating['Movie-Name'].isin(famous_movies)]
```

```python
final_rating=final_rating.merge(poster,on='imdbId')
```

```python
final_rating.drop(['movieId','tmdbId','Movie-Genres','Movie-Name','Imdb Link_y','Title_y','IMDB Score_y','Genre_y','Poster_y'],axis=1)
```

```python
pt = final_rating.pivot_table(index='Title_x',columns='userId',values='rating')
```

```python
pt.fillna(0,inplace = True)
```

```python
pt
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
```

```python
similarity_scores = cosine_similarity(pt)
```

```python
similarity_scores.shape
```

```python
def recommend(movie_name):
    #index fetch
    index = np.where(pt.index== movie_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:6]
    data = []
    
    for i in similar_items:
        item = []
        temp_df = (rating_with_movie[rating_with_movie['Title'] == pt.index[i[0]]])
        item.extend(list(temp_df.drop_duplicates('Title')['Title'].values))
        item.extend(list(temp_df.drop_duplicates('Title')['Genre'].values))
        item.extend(list(temp_df.drop_duplicates('Title')['Poster'].values))
        
        data.append(item)
    return data
        
```

```python
recommend("Young Frankenstein (1974)")
```

```python
import pickle
pickle.dump(pt,open('pt.pkl','wb'))
pickle.dump(rating_with_movie,open('rating_with_movie.pkl','wb'))
pickle.dump(similarity_scores,open('similarity_scores.pkl','wb'))
```
